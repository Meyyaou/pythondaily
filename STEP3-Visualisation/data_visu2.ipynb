```python
#exo
dataset={f"experience{i}": np.random.randn(100,3)for i in range(4)}
#dataset
```

```python
def graph(dataset):
  plt.figure()
  plt.subplot(4,1,1)
  plt.tight_layout()
  plt.plot(dataset['experience0'], c='blue')
  plt.title('exeperience0')

  plt.subplot(4,1,2)
  plt.plot(dataset['experience1'], c='blue')
  plt.title('experience1')

  plt.subplot(4,1,3)
  plt.plot(dataset['experience2'], c='blue')
  plt.title('experience2')

  plt.subplot(4,1,4)
  plt.plot(dataset['experience3'], c='blue')
  plt.title('experience3')

  plt.show()
```

```python
#solu exo plus opt
def graphe(data):
  n=len(data)
  plt.figure(figsize=(12,8))
  #fct zip on fait des iterations sur 2 datastruct en parallele
  for k,i in zip(data.keys(), range(1, n+1)):
    plt.subplot(n, 1, i)
    plt.plot(data[k])
    plt.title(k)
  plt.show()

graphe(dataset)
```

```python
graph(dataset)
```

```python
#scatter classification
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
```

```python
iris= load_iris()

x=iris.data
y=iris.target
names=list(iris.target_names)

print(f'x contient {x.shape[0]} examples et {x.shape[1]} vars')
print(f'il ya {np.unique(y).size} classes')
y
```

```python
#alpha: opacite, s: grosseur du point
plt.scatter(x[:, 0], x[:,1], c=y, alpha=0.5, s=x[:, 2]*100)
plt.xlabel('long sépal')
plt.ylabel('large sépal')
#limit only 2 vars
```

```python
#graphe 3D
from mpl_toolkits.mplot3d import Axes3D
```

```python
ax=plt.axes(projection='3d')
ax.scatter(x[:, 0], x[:, 1], x[:,2], c=y)
```

```python
#visualize des fct
f= lambda x, y: np.sin(x)+np.cos(x+y)*np.cos(x)
```

```python
X= np.linspace(0,5,100)
Y= np.linspace(0,5,100)

X,Y=np.meshgrid(X,Y)
Z=f(X,Y)
```
